```python
import pandas as pd
import os
from underthesea import pos_tag
import itertools
```

```python
#questions + answers
file_names = []
df_list_questions = []
for each in os.listdir():
    if 'xlsx' in each:
        file_names.append(each)
        df = pd.read_excel(each,header=None)
        df_list_questions.append(df)
        print(each)

```

```python
wordnet_location_prefix = '../wordnet-hanhchinh-cong-qn/'
training_questions_prefix = '../cau-hoi-training/'

wordnet_files = []
wordnet_df = []

for each_question_file in os.listdir():
    if 'xlsx' in each_question_file:      
        print(each_question_file)
        for each_wordnet_file in os.listdir(wordnet_location_prefix):
            if 'xlsx' in each_wordnet_file:
                if each_wordnet_file == each_question_file:
                    
                    questions_answer_df = pd.read_excel(each_question_file,header=None).fillna('')
                    wordnet_df = pd.ExcelFile(wordnet_location_prefix+each_wordnet_file)
                    n_sheet = pd.read_excel(wordnet_df,'n',header=None).fillna('')
                    v_sheet = pd.read_excel(wordnet_df,'v',header=None).fillna('')
                    a_sheet = pd.read_excel(wordnet_df,'a',header=None).fillna('')
                    question_list = questions_answer_df[0]
#                     print('báo' in n_sheet[0].values)
                    training_questions = []
                    for each_question in question_list:
                        new_questions = []                        
                        for each_word in pos_tag(each_question):                            
                            word_set = []
                            if each_word[1] == 'N':
                                if each_word[0] in n_sheet[0].values:
#                                     print(n_sheet.loc[n_sheet[0] == each_word[0]].values[0])                                                                        
                                    word_set = [x for x in n_sheet.loc[n_sheet[0] == each_word[0]].values[0].tolist() if x != ''][0:2]
                            elif each_word[1] == 'V':
                                if each_word[0] in v_sheet[0].values:
#                                     print(n_sheet.loc[n_sheet[0] == each_word[0]].values[0])                                                                        
                                    word_set = [x for x in v_sheet.loc[v_sheet[0] == each_word[0]].values[0].tolist() if x != ''][0:2]
                            elif each_word[1] == 'A':
                                if each_word[0] in a_sheet[0].values:                                    
                                    word_set = [x for x in a_sheet.loc[a_sheet[0] == each_word[0]].values[0].tolist() if x != ''][0:2]
#                                     print(a_sheet.x[a_sheet[0] == each_word[0]].values[0])
                            if len(word_set) == 0:
                                word_set.append(each_word[0])                                                            
                            new_questions.append(word_set)
#                         print(new_questions)
                        converted_questions = list(itertools.product(*new_questions))
                        
                        for e in converted_questions:
                            if e != "" :
                                if 'topic1' in each_question_file:
                                    training_questions.append(" ".join(e).replace('"','') + "__label__topic1")
                                elif 'topic2' in each_question_file:
                                    training_questions.append(" ".join(e).replace('"','') + "__label__topic2")
                                elif 'topic3' in each_question_file:
                                    training_questions.append(" ".join(e).replace('"','') + "__label__topic3")  
                                elif 'topic4' in each_question_file:
                                    training_questions.append(" ".join(e).replace('"','') + "__label__topic4")
#                     training_questions_file = pd.Series(training_questions)
#                     training_questions_file.to_csv("training" + each_question_file+".txt",index=False)
                        with open("training" + each_question_file+".txt","w") as f:
                            for each_quest in training_questions:
                                f.write(each_quest)
                        
                        
                        
                        
        

# for each in df_list:
#     questions = each[0].values
#     print('shit')
os.listdir(wordnet_location_prefix)
```
